```python
import torch

a = torch.randn(3, 1)
b = torch.randn(1, 4)
print((a + b).shape)   # → torch.Size([3, 4])

a = torch.randn(5, 1, 3)
b = torch.randn(4, 3)
print((a + b).shape)   # → torch.Size([5, 4, 3])

a = torch.randn(2, 3)
b = torch.randn(3)
print((a * b).shape)   # → torch.Size([2, 3])
```

```python
import torch

x = torch.arange(24)  # shape: (24,)

# reshape
a = x.reshape(4, 6)        # shape: (4, 6)
b = x.reshape(2, 3, 4)     # shape: (2. 3. 4)
c = x.reshape(2, -1)       # shape: (2, 12) (dùng -1 tự tính)

# view (yêu cầu contiguous)
d = x.view(6, 4)           # shape: (6, 4)

# permute
e = torch.randn(2, 3, 4)
f = e.permute(2, 0, 1)     # shape: (4, 2, 3) (đổi thứ tự dims)

# Khi nào view lỗi?
g = e.permute(1, 0, 2)     # non-contiguous!
try:
    h = g.view(3, -1)      # → RuntimeError!
except RuntimeError as err:
    print("❌ view lỗi:", err)
    # Fix:
    h = g.contiguous().view(3, -1) # Them contiguousn de no sap xep lai chieu trong RAM, con permute chi thay doi goc nhin thoi, nen khi su dung view se bi loi~
    print("✅ Fix bằng .contiguous():", h.shape)

print(a.shape, b.shape, c.shape, d.shape, f.shape)
```
